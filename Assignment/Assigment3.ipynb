## Question 1: Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset to avoid errors and display the first 5 rows
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
df.head()
```

### Analysis
The dataset was successfully loaded and the first five rows were displayed.

```python
total_sales_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
total_sales_region
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Category", data=df)
plt.title("Distribution of Orders Across Product Categories")
plt.show()
```

**Analysis:**

The bar chart shows that Office Supplies is the most frequently ordered category, significantly outperforming Furniture and Technology in order volume.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Region", y="Profit", data=df, showfliers=False)
plt.title("Profit Distribution Across Regions")
plt.show()
```

**Analysis:**

The boxplot shows the distribution of profit across the four regions. While the median profit appears relatively similar across all regions, the West and South regions show a slightly higher upper range of profit compared to the Central region, which has a wider spread of lower profit values.

```python
total_profit_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
total_profit_category
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Discount", y="Profit", data=df)
plt.title("Relationship Between Discount and Profit")
plt.show()
```

**Analysis: (Q6)**

The scatterplot reveals a clear negative relationship between Discount and Profit. While low discounts (up to 20%) generally maintain positive profitability, high discount levels (above 0.2) lead to significant financial losses. This suggests that aggressive discounting strategies negatively impact the store's overall profit.
